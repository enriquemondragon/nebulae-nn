Nebulae-NN
===============

**Neural network implementation from scratch for celestial objects classification.**

![nebulae_nn](images/nebulae_nn.png)

## Training

In this section you will be able to train Nebulae-NN using some Hubble's (.jpg) format images available at [NASA](https://www.nasa.gov/content/goddard/hubble-s-messier-catalog#grid)'s webpage. These images have been preprocessed with this [MATLAB script](https://github.com/enriquemondragon/nebulae-nn/blob/main/data_preprocess.m) so that the training can be done in reasonable amount of time. 

To train the model, you will need to provide the training dataset (--input), along with its labels (--labels). I provide those things [here](https://github.com/enriquemondragon/nebulae-nn/tree/main/tutorials_data/train_data).

Some of the things you can tune to train the model are:
| flag | Default value | Description |
| :---: | :---: | :--- |
| --dim_layers | [16,8,1] | List of the layer's dimensions |
| --activation | relu | Activation function for the inner layers. Choices are: [sigmoid, relu, tanh] |
| --alpha | 0.001 | learning rate at which the parameters will be updated |
| --epochs | 2500 | number of complete forward - backward propagation cycles |


```python
! python3 nebulae_nn.py --input tutorials_data/train_data/ --label tutorials_data/train_data/labels_train.csv --dim_layers 16 8 1 --activation relu --alpha 0.001 --epochs 2500
```

After the training Nebulae-NN will save 3 files:
- history.csv: containing the cost per epoch for all the training
- training.png: A figure with the epoch-cost curve
- model.npy: The model
